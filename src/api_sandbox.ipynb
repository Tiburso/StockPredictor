```python
import yfinance as yf
from datetime import date, timedelta, datetime

from helpers.plots import data_plot 

from tickers import TICKERS
```

```python
end_date = date.today().strftime("%Y-%m-%d") #end date for our data retrieval will be current date
start_date = '1990-01-01' # Beginning date for our historical data retrieval

tickers = yf.Tickers(TICKERS) # Create a Tickers object with the tickers we want to retrieve data for
df = tickers.download(start=start_date, end=end_date) # Download the data for the tickers

df.tail() # Display the last 5 rows of the dataframe
```

```python
msft = tickers.tickers["MSFT"] # Get the Microsoft stock from the Tickers object

print(msft.history(period="1mo")) # Get the historical data for the Microsoft stock for the last month
print(msft.info) # Get the information about the Microsoft stock
print(msft.option_chain("2024-03-08")) # Get the option chain for the Microsoft stock
```
